### 1. Wyprowadzenie gradientu funkcji celu ð¹

$$ F(x^{(0)}, x^{(1)}, x^{(2)}, ...x^{(n)}) = Î»_1 \sum_{i=1}^{n}\sum_{j=1}^{k} \frac{1}{Îµ+||x^{(i)} âˆ’r^{(j)}||^2_2} +  Î»_2 \sum_{i=0}^{n-1} ||x^{(i+1)} âˆ’x^{(i)}||^2_2 $$

### Biblioteki

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Podane parametry

```python
n = 20
k = 30
x0 = np.array([0, 0])
xn = np.array([20, 20])
r = np.random.uniform(0, 20, (k, 2))
lambda1 = 1
lambda2 = 1
epsilon = 1e-3
num_iterations = 400
x = np.linspace(x0, xn, n+1)
```

### Funkcja kosztu

- term1: Kara za bliskoÅ›Ä‡ do przeszkÃ³d.
- term2: Kara za dÅ‚ugoÅ›Ä‡ Å›cieÅ¼ki.
- lambda1 i lambda2 rÃ³wnowaÅ¼Ä… te dwie skÅ‚adowe.

```python
def compute_F(x, r, lambda1, lambda2):
    term1 = sum(1 / (np.linalg.norm(x[i] - r[j])**2 + epsilon) for i in range(1, n) for j in range(k))
    term2 = sum(np.linalg.norm(x[i+1] - x[i])**2 for i in range(n))
    return lambda1 * term1 + lambda2 * term2
```

### Gradient
Obliczamy gradient funkcji $ð¹$ wzglÄ™dem kaÅ¼dego punktu $ x^{(i)} $  z wyjÄ…tkiem punktÃ³w koÅ„cowych.

```python
def compute_gradient(x, r, lambda1, lambda2):
    grad = np.zeros_like(x)  
    for i in range(1, n):   
        term1_grad = np.zeros(2)
        for j in range(k):
            term1_grad += (x[i] - r[j]) / (np.linalg.norm(x[i] - r[j])**2 + epsilon)**2
        term1_grad *= -2 * lambda1

        term2_grad = np.zeros(2)
        if i > 0:
            term2_grad += 2 * lambda2 * (x[i] - x[i-1])
        if i < n:
            term2_grad += 2 * lambda2 * (x[i] - x[i+1])

        grad[i] = term1_grad + term2_grad
    return grad
```

### Metoda ZÅ‚otego PodziaÅ‚u do Przeszukiwania Liniowego

Znajdujemy optymalny krok $Î±$ dla najwiÄ™kszego spadku, minimalizujÄ…c funkcjÄ™ kosztu wzdÅ‚uÅ¼ kierunku przeszukiwania.

```python
def golden_section_search(func, a, b, tol=1e-5):
    phi = (1 + np.sqrt(5)) / 2
    while (b - a) > tol:
        c = b - (b - a) / phi
        d = a + (b - a) / phi
        if func(c) < func(d):
            b = d
        else:
            a = c
    return (b + a) / 2
```

### Gradientowy Spadek z Przeszukiwaniem Liniowym
- Iteracyjnie aktualizuje Å›cieÅ¼kÄ™, poruszajÄ…c siÄ™ w kierunku przeciwnym do gradientu.
- UÅ¼ywa przeszukiwania metodÄ… zÅ‚otego podziaÅ‚u do okreÅ›lenia optymalnego kroku $ð›¼$ przy kaÅ¼dej iteracji.


```python
def gradient_descent_with_line_search(x, r, lambda1, lambda2, num_iterations):
    history = []
    for _ in range(num_iterations):
        grad = compute_gradient(x, r, lambda1, lambda2)
        
        # Funkcja do minimalizacji wzdÅ‚uÅ¼ kierunku przeciwnym do gradientu
        def line_search_func(alpha):
            new_x = x.copy()
            new_x[1:n] = new_x[1:n] - alpha * grad[1:n]
            return compute_F(new_x, r, lambda1, lambda2)
        
        alpha = golden_section_search(line_search_func, 0, 1) 
        
        x[1:n] = x[1:n] - alpha * grad[1:n]
        
        history.append(compute_F(x, r, lambda1, lambda2))
    return x, history
```

### 3. NakrÃ³tsza Å›cieÅ¼ka robota

```python
optimal_path, history = gradient_descent_with_line_search(x, r, lambda1, lambda2, num_iterations)
```
